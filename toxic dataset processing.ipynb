```python
import pandas as pd
import numpy as np
```

```python
toxic_original = pd.read_csv("/content/gdrive/MyDrive/Dataset/train.csv").fillna(' ')
```

```python
toxic_original.head()
```

```python
toxic_result = []
for i in toxic_original['target']:
  if i >= 0.5: 
    toxic_result.append(1) # 1 means toxic
  else:
    toxic_result.append(0) # 0 means non-toxic
toxic_original['toxic_result'] = toxic_result
```

```python
toxic_1 = toxic_original['target'] >= 0.5
toxic_df_1 = pd.DataFrame(toxic_original[toxic_1])
toxic_df_1 # check number of toxic text
```

```python
#toxic_df_1 = toxic_df_1.iloc[:15000]
#random select 15000 rows with target(toxic) value >= 0.5
toxic_df_1 = toxic_df_1.sample(n=15000)
toxic_df_1 # select the non toxic text, make it as
```
